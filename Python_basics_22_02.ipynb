Python is strong typed and dynamic typed language
Strong typed means you cannot do operations on different types.
a = 1
b = "2"
a + b will result in a TypeError
TypeError: unsupported operand type(s) for +: 'int' and 'str'

Dynamic typed language means you can assign different types to the same variable
a = 1
a = "Hello World"
is a valid code.

**Naming conventions**  
Variable and definition(functions) names are separated by "_"  
```
lst_my_list = []
def my_function():  
    print ("my beautiful function")
```
For the class names each word is capitalized:  
class MyBeautifulClass():pass

Definition of various data types.  
It's a good practice to prepand the variable type:  
lst_my_values : a list(array)  
dct_my_values: dictionary(obj class)

etc..

**Indentation**  
In python indentation is **everything**  
The interpreter is relying on the indentation to figure out where a function, loop, class etc starts and ends. Default indentation is **4 spaces**   
Most IDEs are using the TAB to create those spaces

````
#definition declaration
def my_simple_function(condition=False):
    if condition:
        print("hey")
    print("hello")

my_simple_function(True)
print("no more in my simple function")
````

**Definition delcaration**  
The syntax is:  


```
def my_simple_function(mendatory_arg, optional_arg=None, *args, **kwargs):
    print("Python is amazing")
```
**Arguments:**  
  
**mandatory_arg**:  
this argument is mendatory, when calling the function. Otherwise the interpreter will trow an error  
**optional_arg**:  
every argument with a default value in the function declaration is optional.
If no argument passed, the function will use the default value.  
***args**: a list placeholder for additional arguments. You don't have to declare every argument in the function. If for some reason you need more arguments, you can pass them to the function and get them from the args parameter.  
****kwargs**: a dictionary placeholder for additional arguments. The same as the
args but in key value form.  

**The order of the arguments is importart**  
Firs the mandatory, then the optional, then the args then the kwargs.
my_simple_function(optional_arg=5, mendatory_arg) will trow an error.




**Code complexity**  
The code complexity(big O) is used to determine how many computations are needed to accomplish the goal.  
For example an iteration over all values of an array with n-elements will have  
the complexity of O(n)

```
lst = [1, 2, 3, 4, 5, 6, 7]
for item in lst:
    print(item)
```

checking if a key in dictionary exists has the complexity of O(1).

```
dct = {
    "smallest_item": 1,
    "largest_item": 69
}
if "smallest_item" in dct:
    print(dct["smallest_item"])
```
**O(1)** complexity is the dream, but rarely the case.** O(n)** is fine, everything higher should be avoided. **O(nÂ²)** is a disaster.
